## Recuperar el n√∫mero de orden del cliente

El reclamo puede comenzar de esta forma:

Seguido, a que a su vez el chatbot le pida su n√∫mero de telefono y el mail asociado al mismo:

Que podr√≠a incluir todos estos formatos:

```python
import re

chat1 = 'codebasics: you ask lot of questions üò†  1235678912, abc@xyz.com'
chat2 = 'codebasics: here it is: (123)-567-8912, abc_x98@xyz.com'
chat3 = 'codebasics: yes, phone: 1235678912 email: abc@xyz.edu.ar'

pattern = '\d{10}|\(\d{3}\)-\d{3}-\d{4}'

busqueda = re.findall (pattern, chat3)
busqueda
```

```python
pattern_mail = '[a-z0-9A-Z_]*@[a-z0-9A-Z]*.[a-zA-Z]*.[a-z]*'

busqueda_mail = re.findall(pattern_mail, chat3)
busqueda_mail
```

```python
import re

chat1='codebasics: Hello, I am having an issue with my order # 412889912'

pattern = 'order[^\d]*(\d*)'
matches = re.findall(pattern, chat1)
matches
```

```python
chat2='codebasics: I have a problem with my order number 412889912'
pattern = 'order[^\d]*(\d*)'
matches = re.findall(pattern, chat2)
matches
```

```python
chat3='codebasics: My order 412889912 is having an issue, I was charged 300$ when online it says 280$'
pattern = 'order[^\d]*(\d*)'
matches = re.findall(pattern, chat3)
matches
```

```python
def get_pattern_match(pattern, text):
    matches = re.findall(pattern, text)
    if matches:
        return matches[0]
```

```python
get_pattern_match('order[^\d]*(\d*)', chat1)
```

```python
def extract_personal_information(text):
    age = get_pattern_match('age (\d+)', text)
    full_name = get_pattern_match('Born(.*)\n', text)
    birth_date = get_pattern_match('Born.*\n(.*)\(age', text)
    birth_place = get_pattern_match('\(age.*\n(.*)', text)
    return {
        'age': int(age),
        'name': full_name.strip(),
        'birth_date': birth_date.strip(),
        'birth_place': birth_place.strip()
    }
```

```python
extract_personal_information(text)
```
